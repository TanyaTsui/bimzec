```python
import matplotlib.pyplot as plt
from matplotlib.ticker import ScalarFormatter

data1 = {0: 328534, 1: 413923, 2: 528496}
data2 = {0: 985604, 1: 1241770, 2: 1585490}
data3 = {0: 109511, 1: 137974, 2: 176165}

keys = list(data1.keys())
values1 = list(data1.values())
values2 = list(data2.values())
values3 = list(data3.values())

bar_width = 0.3 # Width of each bar
index = range(len(keys))  # X-axis index for positioning bars

plt.bar(index, values1, width=bar_width, label='Train')
plt.bar([i + bar_width for i in index], values2, width=bar_width, label='Truck')
plt.bar([i + 2 * bar_width for i in index], values3, width=bar_width, label='Ship')

plt.gca().yaxis.set_major_formatter(ScalarFormatter(useMathText=True))
plt.gca().ticklabel_format(style="plain", axis="y")  # Disable scientific notation

plt.xlabel("Construction Projects")
plt.ylabel("CO2 in Kg")
plt.title("CO2 Emission Comparsion")
plt.xticks([i + bar_width/2 for i in index], keys)  # Center x-ticks between grouped bars
plt.legend()
plt.show()
```
