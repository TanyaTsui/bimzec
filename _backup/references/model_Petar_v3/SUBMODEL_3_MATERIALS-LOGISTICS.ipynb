```python

df = materials_logistics

# Create a bar plot
plt.figure(figsize=(9,6))
plt.bar(df['material'], df['multiplication_factor'], color='b')
plt.xlabel('Material')
plt.ylabel('Multiplication Factor')
plt.title('Multiplication Factor for Materials')
plt.xticks(rotation=45, ha='right')

plt.tight_layout()
plt.show()
```
