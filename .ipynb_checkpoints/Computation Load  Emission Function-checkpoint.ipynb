```python
#It is based on a paper by Marcel Turkensteen "The computation of carbon emissions due to the net payload on a truck"
#There are different studies on the fuel consumtion / emission difference for diverse freight vehicles by diverse load factor'''
```

```python
from IPython.display import Image


# Display the image in the notebook
Image(filename='loademission.png')
```

```python
def calculate_new_emissionCO2(full_emissionCO2, difference, full_weight, empty_weight, current_weight):
    empty_emissionCO2 = full_emissionCO2 * difference
    emission_difference = full_emissionCO2 - empty_emissionCO2
    difference_weight = full_weight - empty_weight
    emission_diff_tone = emission_difference / difference_weight
    new_emission_top = emission_diff_tone * (current_weight - empty_weight)
    new_emissionCO2 = empty_emissionCO2 + new_emission_top
    return new_emissionCO2

# Example usage:
result = calculate_new_emissionCO2(1539, 0.66, 50, 24, 46)
print(result)
```
